```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hr_data= pd.read_csv(r"F:\MeriSkill Internship\HR Dataset\Data P3 MeriSKILL\HR-Employee-Attrition.csv")
```

```python
hr_data.info()
```

## Tasks to perform:

```python
hr_data.head(2)
```

```python
# Deleting redundant columns.
hr_data.drop(columns= ['StockOptionLevel', 'DailyRate', 'TrainingTimesLastYear', 
                       'YearsSinceLastPromotion', 'MonthlyRate', 'JobInvolvement', 
                       'HourlyRate'], inplace= True)
```

```python
hr_data.columns
```

```python
# Renaming the columns.
hr_data.rename(columns= {'BusinessTravel': 'Business_Travel', 'DistanceFromHome': 'Distance_From_Home', 
                         'EducationField': 'Education_Field', 'EmployeeCount': 'Employee_Count', 
                         'EmployeeNumber': 'Employee_Number',  'EnvironmentSatisfaction': 'Environment_Satisfaction', 
                        'JobLevel': 'Job_Level', 'JobRole': 'Job_Role', 'JobSatisfaction': 'Job_Satisfaction' ,
                         'MaritalStatus': 'Marital_Status', 'MonthlyIncome': 'Monthly_Income', 
                         'NumCompaniesWorked': 'Num_Companies_Worked', 'Over18': 'Over_18', 'OverTime': 'Over_Time',
                         'PercentSalaryHike': 'Percent_Salary_Hike', 'PerformanceRating': 'Performance_Rating',
                         'RelationshipSatisfaction': 'Relationship_Satisfaction', 'StandardHours': 'Standard_Hours',
                         'TotalWorkingYears': 'Total_Working_Years', 'WorkLifeBalance': 'Work_Life_Balance', 
                         'YearsAtCompany': 'Years_At_Company', 'YearsInCurrentRole': 'Years_In_Current_Role',
                         'YearsWithCurrManager': 'Years_With_Curr_Manager' }, inplace= True )
```

```python
# Dropping duplicates.
hr_data.drop_duplicates(inplace= True)
```

```python
# Cleaning individual columns.
All the columns are already clean.
```

```python
# Remove the NaN values from the dataset
hr_data.isna().sum()
```

```python
No Null values to be found in the dataset.
```

```python
# Check for some more Transformations
hr_data['Attrition'] = np.where(hr_data.Attrition == 'Yes', 1, 0)
hr_data['Over_Time']= np.where(hr_data.Over_Time == 'Yes', 1, 0)
```

## Data Visualisations
